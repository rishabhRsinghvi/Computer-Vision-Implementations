```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
import numpy as np
import random
```

```python
img = cv2.imread('/content/school.jpg',0)
cv2_imshow(img)
```

**Basic High Pass Filtering**

```python
row,col = img.shape
```

```python
mask = np.zeros([3, 3], dtype = int)
for i in range(3):
  for j in range(3):
    if i==1 and j==1:
      mask[i,j] = 8
    else:
      mask[i,j]=-1
```

```python
mask
```

```python
img_new = np.zeros([row, col])
```

```python
for i in range(1, row-1):
    for j in range(1, col-1):
        temp = img[i-1, j-1]*mask[0, 0]+img[i-1, j]*mask[0, 1]+img[i-1, j + 1]*mask[0, 2]+img[i, j-1]*mask[1, 0]+ img[i, j]*mask[1, 1]+img[i, j + 1]*mask[1, 2]+img[i + 1, j-1]*mask[2, 0]+img[i + 1, j]*mask[2, 1]+img[i + 1, j + 1]*mask[2, 2]
        img_new[i, j]= temp
```

```python
cv2_imshow(img_new)
```

**High Boost Filtering**

```python
img = cv2.imread('/content/school.jpg',0)
cv2_imshow(img)
```

```python
row,col = img.shape
```

```python
A=1
```

```python
mask = np.zeros([3, 3], dtype = int)
for i in range(3):
  for j in range(3):
    if i==1 and j==1:
      mask[i,j] = A+8
    else:
      mask[i,j]=-1
```

```python
img_new = np.zeros([row, col])
```

```python
for i in range(1, row-1):
    for j in range(1, col-1):
        temp = img[i-1, j-1]*mask[0, 0]+img[i-1, j]*mask[0, 1]+img[i-1, j + 1]*mask[0, 2]+img[i, j-1]*mask[1, 0]+ img[i, j]*mask[1, 1]+img[i, j + 1]*mask[1, 2]+img[i + 1, j-1]*mask[2, 0]+img[i + 1, j]*mask[2, 1]+img[i + 1, j + 1]*mask[2, 2]
        img_new[i, j]= temp
```

```python
cv2_imshow(img_new)
```
